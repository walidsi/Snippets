```python
import requests
import json
import pandas as pd
```

```python
df = pd.read_excel('Australian Post Codes.xlsx')
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
lat = (df['Latitude'].max() - df['Latitude'].min())/2 + df['Latitude'].min()
long = (df['Longitude'].max() - 75)/2 + 75

print((lat, long))
```

```python
df['Longitude'].min()
```

```python
grouped = df.groupby('Postcode')
```

```python
import folium
```

```python
m = folium.Map(location=[lat, long], zoom_start=4)

for name, group in grouped:
    #print(name)
    
    marker = folium.CircleMarker(
            location = [group.Latitude.mean(), group.Longitude.mean()],
            radius = 5,
            fill_opacity = 0.8, color=None,
            fill_color = ('#274cc9'),
            popup = 'Number of people: {}'.format(group.shape[0]),
            tooltip = 'Postcode: {}, Lat:{}, Long:{}<br>Click for more info.'.format(name, group.Latitude.mean(), group.Longitude.mean())
            )

    marker.add_to(m) 

m
```
